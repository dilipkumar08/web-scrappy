```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/SpaceX'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

```python
soup.find('table',class_='wikitable plainrowheaders')  
```

```python
table=soup.find_all('table')[3]
```

```python
table.find_all('th')
```

```python
columns=[]
for title in table.find_all('th'):
    print(title.text.strip())
    columns.append(title.text.strip())
```

```python
print(columns)
```

```python
data=table.find_all('tr')
rows=[]
for row in data[1:]:
    row_data=row.find_all('td')
    rows.append([rowdata.text.strip() for rowdata in row_data])
```

```python
print(rows)
```

```python
achievements_summary=pd.DataFrame(rows,columns=columns)
```

```python
achievements_summary
```
